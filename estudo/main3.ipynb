```python
import pandas as pd
import numpy as np
from pandas_datareader import data
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


import yfinance as yf
yf.pdr_override()
```

```python
acoes = [
  
    "VALE3.SA",
    "RECV3.SA",
    "KEPL3.SA",
    "CURY3.SA",
    "VULC3.SA",
    "LEVE3.SA",
    "GGBR4.SA",
    "LREN3.SA",
    "ALSO3.SA",
    "DIRR3.SA",
    "MILS3.SA",
    "RANI3.SA",
    "CAMB3.SA",
    "CSUD3.SA",
    "CLSC4.SA",
    "CMIG4.SA",
    "ITSA4.SA",
    "BBAS3.SA",
    "TASA4.SA",
    "SLCE3.SA",
    "B3SA3.SA",
    "PETR4.SA",
    "SUZB3.SA",
    "KLBN4.SA",
    "BOVA11.SA"

        
]  


acoes_df1 = pd.DataFrame()
for acao in acoes:
    acoes_df1[acao] = data.get_data_yahoo(acao, start='2010-08-01')['Close']
acoes_df1    
```

```python
acoes_df1.dropna(inplace=True)
acoes_df1.to_csv("db/acoes1.csv")
acao_df1_normalizado = acoes_df1.copy()
for i in acao_df1_normalizado.columns[0:]:
  acao_df1_normalizado[i] = acao_df1_normalizado[i] / acao_df1_normalizado[i][0]

acao_df1_normalizado
```

```python
empresas = ["VALE3.SA", "RECV3.SA", "KEPL3.SA", "CURY3.SA", "VULC3.SA", "LEVE3.SA", "GGBR4.SA", "LREN3.SA", "ALSO3.SA", "DIRR3.SA", "MILS3.SA", "RANI3.SA", "CAMB3.SA", "CSUD3.SA", "CLSC4.SA", "CMIG4.SA", "ITSA4.SA", "BBAS3.SA", "TASA4.SA", "SLCE3.SA", "B3SA3.SA", "PETR4.SA", "SUZB3.SA", "KLBN4.SA"]
pesos_carteira1 = np.array([0.038, 0.024, 0.0195, 0.0285, 0.0285, 0.0195, 0.0195,0.0195,0.015,0.02,0.0195,0.0195,0.004,0.02,0.02,0.0285,0.024,0.02,0.012,0.0185,0.019,0.029,0.017,0.017])
soma_valores_acoes = acao_df1_normalizado[empresas].multiply(2*pesos_carteira1, axis=1)
soma_valores_acoes = soma_valores_acoes[empresas].sum(axis=1)
acao_df1_normalizado['Carteira'] = soma_valores_acoes
acao_df1_normalizado
```

```python
figura = px.line(title = 'Comparativo carteira x BOVA')
acao_df1_normalizado.drop(empresas, axis=1, inplace=True)
acao_df1_normalizado.reset_index(inplace=True)
for i in acao_df1_normalizado.columns[1:]:
  figura.add_scatter(x = acao_df1_normalizado['Date'], y = acao_df1_normalizado[i], name = i)
figura.show()
```
